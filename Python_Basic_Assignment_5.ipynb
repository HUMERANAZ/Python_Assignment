#### 1. What does an empty dictionary's code look like?

An empty dictionary's code representation is `{}`. It consists of a pair of curly braces with no key-value pairs inside.



#### 2. What is the value of a dictionary value with the key 'foo'and the value 42?

The value of a dictionary with the key `'foo'` and the value `42` would be `42`. In other words, accessing `spam['foo']` would return `42` if `spam` is the dictionary `{ 'foo': 42 }`.




#### 3. What is the most significant distinction between a dictionary and a list?

The most significant distinction between a dictionary and a list is their structure and usage. A dictionary is an unordered collection of key-value pairs, where each key is unique and associated with a value. On the other hand, a list is an ordered collection of values that can be accessed by their position or index.

In a dictionary, values are accessed by their corresponding keys, allowing for efficient lookup and retrieval of values based on a specific identifier. In contrast, a list is accessed by its index, which represents the position of the element in the list. Lists are commonly used when the order of elements matters or when there is a need for sequential data storage.




#### 4. What happens if you try to access spam['foo'] if spam is {'bar': 100}?

If you try to access `spam['foo']` where `spam` is `{ 'bar': 100 }`, it will raise a `KeyError`. This is because the key `'foo'` does not exist in the dictionary `spam`. Accessing a non-existent key in a dictionary results in an error.




#### 5. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.keys()?

The expression `'cat' in spam` checks if the key `'cat'` exists in the dictionary `spam`. It returns `True` if the key is present and `False` otherwise.

On the other hand, the expression `'cat' in spam.keys()` also checks if the key `'cat'` exists in the dictionary `spam`. It returns the same result as `'cat' in spam`.




#### 6. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.values()?

The expression `'cat' in spam` checks if the value `'cat'` exists as a value in the dictionary `spam`. It returns `True` if the value is present and `False` otherwise.

In contrast, the expression `'cat' in spam.values()` checks if the value `'cat'` exists among the values in the dictionary `spam`. It returns `True` if the value is present and `False` otherwise.




#### 7. What is a shortcut for the following code? 
#### if 'color' not in spam:
#### spam['color'] = 'black'

A shortcut for the code snippet you provided can be achieved using the `dict.setdefault()` method. It sets the value for a specified key if the key is not already present in the dictionary.

Shortcut code:
```python
spam.setdefault('color', 'black')
```

This line of code checks if the key `'color'` exists in the `spam` dictionary. If the key is not present, it adds it with the value `'black'`. If the key already exists, it does nothing.




#### 8. How do you "pretty print" dictionary values using which module and function?

To "pretty print" dictionary values, you can use the `pprint` module and its `pprint()` function. The `pprint` module provides a way to display data structures in a more human-readable format, especially for complex or nested data like dictionaries.

Example usage:
```python
import pprint

spam = {'key1': 'value1', 'key2': 'value2'}
pprint.pprint(spam)
```

The `pprint.pprint()` function will print the dictionary `spam` in a more visually organized and indented format, making it easier to read and understand the structure of the dictionary.
