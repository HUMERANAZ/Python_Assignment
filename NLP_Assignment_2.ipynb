#### 1. What are Corpora?

Corpora, also known as a corpus (plural: corpora), refer to large and structured collections of texts or linguistic data. They are used in natural language processing (NLP) and computational linguistics for various purposes, such as training and evaluating language models, conducting linguistic research, and developing language processing algorithms.




#### 2. What are Tokens?

Tokens are individual units or elements that make up a text. In the context of NLP, tokens typically represent words or characters. Tokenization is the process of splitting a text into individual tokens, which serves as the first step in many NLP tasks.




#### 3. What are Unigrams, Bigrams, Trigrams?

Unigrams, bigrams, and trigrams are different types of n-grams, which are contiguous sequences of n items from a given text. 

- Unigrams: Unigrams are n-grams of size one, representing single words. For example, in the sentence "I love to eat pizza," the unigrams are "I," "love," "to," "eat," and "pizza."

- Bigrams: Bigrams are n-grams of size two, representing pairs of consecutive words. For example, in the same sentence, the bigrams are "I love," "love to," "to eat," and "eat pizza."

- Trigrams: Trigrams are n-grams of size three, representing triplets of consecutive words. In the example sentence, the trigrams are "I love to," "love to eat," and "to eat pizza."





#### 4. How to generate n-grams from text?

To generate n-grams from text, you can follow these steps:

- Tokenize the text into individual words or characters.
- Slide a window of size n across the tokens, where n represents the desired number of tokens in each n-gram.
- Collect the n-gram sequences generated by the window as you move across the tokens.

For example, to generate trigrams from the sentence "I love to eat pizza," you would tokenize the sentence into individual words and then slide a window of size three across the words to collect the trigrams.




#### 5. Explain Lemmatization

Lemmatization is the process of reducing words to their base or dictionary form, known as a lemma. The lemma represents the canonical form of a word, regardless of its inflected or derived variations. 

For example, the lemma of the words "running," "ran," and "runs" would be "run." Lemmatization takes into account the word's part of speech and applies linguistic rules to transform it to its base form. It helps reduce word variants and allows for more accurate analysis and comparison of texts.





#### 6.  Explain Stemming

Stemming is a process of reducing words to their base form by removing suffixes or prefixes. Unlike lemmatization, stemming does not consider the word's part of speech or apply linguistic rules but rather applies heuristics and pattern-based rules to chop off affixes.

For example, stemming would convert the words "running," "ran," and "runs" to the stem "run." While stemming is a simpler and faster process compared to lemmatization, it may produce stems that are not actual words, as it focuses on pattern matching rather than linguistic accuracy.





#### 7.  Explain Part-of-speech (POS) tagging

Part-of-speech (POS) tagging is the process of assigning grammatical tags to each word in a given text, indicating their respective part of speech (e.g., noun, verb, adjective, etc.). POS tagging helps in understanding the syntactic structure of a sentence, disambiguating word meanings, and enabling more accurate language processing tasks.

For example, in the sentence "The cat is sleeping," a POS tagger would assign the tag "DT" (determiner) to "The," "NN" (noun) to "cat," "VBZ" (verb) to "is," and "VBG" (verb) to "

sleeping."





#### 8. Explain Chunking or shallow parsing

Chunking, also known as shallow parsing, is the process of grouping or chunking together words in a sentence based on their syntactic structure. It involves identifying and extracting phrases or chunks that serve specific grammatical roles, such as noun phrases (NP) or verb phrases (VP).




#### 9. Explain Noun Phrase (NP) chunking

Noun Phrase (NP) chunking is a specific type of chunking that focuses on identifying and extracting noun phrases from a sentence. A noun phrase is a group of words centered around a noun that function together as a single unit.

For example, in the sentence "The big brown dog chased the cat," the NP chunks would be "The big brown dog" and "the cat."





#### 10. Explain Named Entity Recognition

Named Entity Recognition (NER) is a natural language processing task that involves identifying and classifying named entities, such as person names, locations, organizations, dates, and other proper nouns, within a given text. NER helps in extracting structured information from unstructured text and is used in various applications, including information extraction, question answering systems, and text summarization.

For example, in the sentence "Apple Inc. is planning to open a new store in New York City," NER would identify "Apple Inc." as an organization and "New York City" as a location.
