#### Q5. Compare and contrast the three types of array multiplication between two numpy arrays: dot product, outer product, and regular multiplication of two numpy arrays.

The three types of array multiplication between two NumPy arrays are:

- Dot product: The dot product is performed using the `numpy.dot()` function or the `@` operator. It computes the sum of the element-wise products of corresponding elements in the arrays.

- Outer product: The outer product is calculated using the `numpy.outer()` function. It produces a new array where each element is the product of every combination of elements from the input arrays.

- Regular multiplication: Regular multiplication of two NumPy arrays is performed using the `*` operator. It performs element-wise multiplication, resulting in an array with the same shape as the input arrays.

In summary, the dot product yields a scalar or a lower-dimensional array, the outer product generates a higher-dimensional array, and regular multiplication performs element-wise multiplication.

#### Q7. Can string data be stored in numpy arrays? If so, list at least one restriction that applies to this data.

Yes, string data can be stored in NumPy arrays. However, there are some restrictions to keep in mind:

- All elements in a NumPy array must have the same data type. Therefore, if you store string data in a NumPy array, all the strings must have the same length to ensure a fixed-size representation.

- NumPy arrays are optimized for numerical computations, so string operations may not be as efficient or flexible as with other data structures such as Python lists or pandas DataFrames.

- The maximum length of the strings in a NumPy array is determined by the maximum length specified when creating the array. If a string exceeds the maximum length, it will be truncated.



- String operations in NumPy are generally limited to basic element-wise operations or simple string manipulation functions. More advanced string operations are better suited for other libraries or built-in Python string methods.

It's worth noting that if you need to work extensively with string data, libraries such as pandas provide more comprehensive support and functionality for handling and analyzing string data in tabular structures.
