#### 1. What advantages do Excel spreadsheets have over CSV spreadsheets?

Excel spreadsheets have several advantages over CSV (Comma-Separated Values) spreadsheets:

- Excel spreadsheets can contain multiple sheets within a single file, allowing for better organization and structuring of data. CSV files, on the other hand, typically represent only one sheet of data per file.
- Excel spreadsheets can have complex formatting options, including cell formatting, font styles, colors, and formulas. CSV files, being plain text files, do not support such formatting options.
- Excel spreadsheets can contain charts, graphs, and other visual elements, making them more suitable for data visualization. CSV files only contain raw data without any visual representation.
- Excel spreadsheets can have password protection and access controls, allowing for secure storage of sensitive data. CSV files do not have built-in security features.



#### 2. What do you pass to csv.reader() and csv.writer() to create reader and writer objects?

To create reader and writer objects using the `csv.reader()` and `csv.writer()` functions, you need to pass a File object as an argument.

Example:
```python
import csv

# Creating a reader object
with open('data.csv', 'r') as file:
    reader = csv.reader(file)

# Creating a writer object
with open('output.csv', 'w') as file:
    writer = csv.writer(file)
```

In the above code, `csv.reader()` is called with the `file` object opened in read mode (`'r'`), and `csv.writer()` is called with the `file` object opened in write mode (`'w'`).




#### 3. What modes do File objects for reader and writer objects need to be opened in?

File objects for reader and writer objects need to be opened in the following modes:

- Reader object: The file should be opened in text mode with `'r'` or `'rt'` mode. For example, `open('data.csv', 'r')`.
- Writer object: The file should be opened in text mode with `'w'` or `'wt'` mode. For example, `open('output.csv', 'w')`.




#### 4. What method takes a list argument and writes it to a CSV file?

The `writerow()` method is used to write a list argument to a CSV file. It belongs to the writer object created using `csv.writer()`.

Example:
```python
import csv

data = ['John', 'Doe', 'john.doe@example.com']

with open('output.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(data)
```

In the above code, `writer.writerow(data)` writes the elements of the `data` list as a row in the CSV file.





#### 5.  What do the keyword arguments delimiter and line terminator do?

The `delimiter` and `lineterminator` are keyword arguments used in CSV operations:

- `delimiter`: It specifies the character used to separate fields in a CSV file. By default, it is set to `','` (comma), but it can be changed to any other character that suits the CSV file format.
- `lineterminator`: It specifies the character used to terminate lines in a CSV file. By default, it is set to `'\r\n'` (CR+LF), which is the standard line terminator in CSV files. However, it can be customized as per the requirements of the CSV file format.





#### 6. What function takes a string of JSON data and returns a Python data structure?

The `json.loads()` function takes a string of JSON data and returns a Python data structure. It is used to parse and convert JSON data into a corresponding Python object.

Example:
```python
import json

json_data = '{"name": "John", "age": 30, "city": "New York"}'
python_data = json.loads(json_data)

print(python_data)
```

In the above code, `json.loads(json_data)` converts the JSON string `json_data` into a Python data structure, which is then stored in the `python_data` variable.




#### 7. What function takes a Python data structure and returns a string of JSON data?

The `json.dumps()` function takes a Python data structure and returns a string of JSON data. It is used to serialize a Python object into its corresponding JSON representation.

Example:
```python
import json

python_data = {"name": "John", "age": 30, "city": "New York"}
json_data = json.dumps(python_data)

print(json_data)
```

In the above code, `json.dumps(python_data)` serializes the `python_data` dictionary into a JSON string, which is then stored in the `json_data` variable.
