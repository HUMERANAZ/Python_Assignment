#### Q1.  What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?

A probability distribution is a mathematical function or a table that describes the likelihood of different outcomes or values occurring in a particular event or random experiment. It provides information about the probabilities associated with each possible outcome.

While the values in a probability distribution are meant to represent random events or variables, the distribution itself provides a framework for understanding the likelihood of those events or variables occurring. It does not predict specific outcomes but rather gives insights into the overall pattern or behavior of the random process based on probabilities.




#### Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?

True random numbers are generated from physical processes that are inherently unpredictable, such as atmospheric noise or radioactive decay. They are considered genuinely random as they are not influenced by any deterministic algorithm or formula.

On the other hand, pseudo-random numbers are generated by algorithms that use a seed value as input. While the generated sequence appears random, it is actually deterministic and reproducible given the same seed value. Pseudo-random numbers are considered "good enough" for many applications because they exhibit statistical properties similar to true random numbers and are computationally efficient to generate.




#### Q3. What are the two main factors that influence the behaviour of a &quot;normal&quot; probability distribution?

The two main factors that influence the behavior of a "normal" probability distribution are the mean (average) and the standard deviation. The mean determines the central tendency of the distribution, indicating where the peak or center of the distribution is located. The standard deviation controls the spread or dispersion of the distribution, indicating how much the values deviate from the mean.




#### Q4. Provide a real-life example of a normal distribution.


A real-life example of a normal distribution is the distribution of human heights. In a large population, heights tend to follow a bell-shaped curve with the majority of individuals clustered around the average height, while fewer individuals are found at extreme heights.




#### Q5.  In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?


In the short term, the behavior of a probability distribution can be unpredictable, and individual outcomes may deviate significantly from the expected probabilities. However, as the number of trials increases, the distribution tends to stabilize and converge toward the expected probabilities. This is known as the law of large numbers, which states that the average of a large number of independent and identically distributed random variables will approach the expected value.

#### Q7. Describe the math package&#39;s general categories of functions.


The math package in Python provides several general categories of functions:

- Mathematical constants: Constants such as pi and e.
- Arithmetic functions: Functions for basic mathematical operations like addition, subtraction, multiplication, division, etc.
- Power and logarithmic functions: Functions for exponentiation, logarithms, and related operations.
- Trigonometric functions: Functions for sine, cosine, tangent, and other trigonometric operations.
- Hyperbolic functions: Functions for hyperbolic sine, cosine, tangent, etc.
- Miscellaneous functions: Functions for rounding, absolute value, factorial, etc.




#### Q8.  What is the relationship between exponentiation and logarithms?


Exponentiation and logarithms are inverse operations of each other. Exponentiation involves raising a base number to a power, while logarithms determine the exponent to which a base number must be raised to obtain a given value.

For example, in the equation `b^x = y`, exponentiation is used to find the value of `y` given the base `b` and exponent `x`. Conversely, in the equation `log_b(y) = x`, logarithms are used to find the value of `x` given the base `b` and value `y`.





#### Q9. What are the three logarithmic functions that Python supports?

Python supports three logarithmic functions:

- `math.log(x)`: Computes the natural logarithm (base e) of a number `x`.
- `math.log10(x)`: Computes the base-10 logarithm of a number `x`.
- `math.log(x, base)`: Computes the logarithm of a number `x` with a specified base.
